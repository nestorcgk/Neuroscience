```python
import plotly.plotly as py
import plotly.graph_objs as go
import pandas as pd
import numpy as np
```

```python
data_sch = pd.read_csv('Data_scheduling.csv')
```

```python
data_sch
```

```python
colours = {
    2: 'rgb (178,24,43)',
    3: 'rgb (214,96,77)',
    4: 'rgb (244,165,130)',
    5: 'rgb (253,219,199)',
    6: 'rgb (209,229,240)',
    7: 'rgb (146,197,222)',
    8: 'rgb (67,147,195)',
    9: 'rgb (33,102,172)',
}
```

```python
colours[str(list(data_sch['worker'])[0])]
```

```python
traces = []
x_ = list(data_sch['frame number'])
y_ = list(data_sch['elapsed'])
colours_ = list(data_sch['worker'])

for i in range(len(x_)):
    # Create traces
    trace = go.Scatter(
        x = x_[i],
        y = y_[i],
        name = 'worker:' + str(colours_[i]),
        mode = 'markers',
        marker = dict(color = colours[colours_[i]])
    )
    traces.append(trace)

layout = dict(title = 'Tiempo de procesamiento por muestra ordenada en el tiempo',
              xaxis = dict(title = 'número de cuadro'),
              yaxis = dict(title = 'tiempo de procesamiento'),
              )
data = traces
fig = dict(data=data,layout=layout)
py.iplot(fig, filename='scatter-mode')
```
