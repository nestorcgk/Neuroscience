```julia
K = readdlm("K4096S.dat")
```

```julia
using PyPlot
```

```julia
ConjuntoDeCoordenadasTotal=Array[]
```

```julia
for j=1:64,k=1:64
    push!(ConjuntoDeCoordenadasTotal,[j,k])
end
```

```julia
jmax=4096

for j=1:jmax
    xj=ConjuntoDeCoordenadasTotal[j]
    for k=j+1:jmax
	    xk=ConjuntoDeCoordenadasTotal[k]  
        K[j,k] = K [k,j]
    end
end
```

```julia
matshow(K)
```

```julia
addprocs(8)
```

```julia
kInv = inv(K)
```

```julia
writedlm("kInvSuave.dat",kInv)
```
