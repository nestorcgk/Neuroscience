```julia
for j=1:64,k=1:64
    push!(ConjuntoDeCoordenadasTotal,[j,k])
end
```

```julia
jmax=4096

for j=1:jmax
    xj=ConjuntoDeCoordenadasTotal[j]
    for k=j+1:jmax
	    xk=ConjuntoDeCoordenadasTotal[k]  
        output4096[j,k] = 0
    end
end
```
